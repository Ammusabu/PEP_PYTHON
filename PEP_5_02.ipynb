**SQLITE**

```python
import sqlite3
connection1 = sqlite3.connect("school.db") #connect() will open database or create database
print("Database Connected")
connection1.close()
```

```python
#cursor() will run sql
connection1 = sqlite3.connect("school.db")
cursor1 = connection1.cursor()
print("Cursor Created")
connection1.close()
```

```python
from sqlite3.dbapi2 import connect
#execute() will execute sql
connection1 = sqlite3.connect("school.db")
cursor1 = connection1.cursor()
cursor1.execute("CREATE TABLE IF NOT EXISTS Student(id int primary key,name text,marks int)")
connection1.commit()
print("Table Created")
connection1.close()
```

```python
connection1 = sqlite3.connect("school.db")
cursor1 = connection1.cursor()
cursor1.execute('INSERT INTO Student VALUES(1,"John",90), (2,"Diya",90)')
connection1.commit()
print("Data Inserted")
connection1.close()
```

```python
connection1 = sqlite3.connect("school.db")
cursor1 = connection1.cursor()
cursor1.execute('select * from student')
rows = cursor1.fetchall()
for i in rows:
  print(i)
connection1.close()
```

```python
connection2 = sqlite3.connect("school.db")
cursor2 = connection2.cursor()
cursor2.execute("CREATE TABLE IF NOT EXISTS employee(id int primary key,name text,salary int)")
connection2.commit()
print("Table employee Created")
connection2.close()
```
